```python
import pandas as pd
df_cour = pd.read_csv("coursera_courses.csv")
df_cour_rev = pd.read_csv("coursera_reviews.csv")
```

```python
df_cour.head()
```

```python
#merge two dataframes
df_cour= pd.merge(df_cour_rev,df_cour,on="course_id", how= 'left')
```

```python
df_cour.head()
df_cour.drop(columns=["Unnamed: 0_x","Unnamed: 0_y"],inplace=True)
```
